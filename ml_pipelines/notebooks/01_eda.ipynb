# Exploratory Data Analysis - Land Intelligence Platform


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/sample_land_data.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
df['market_price'].hist(ax=axes[0,0], bins=50)
axes[0,0].set_title('Market Price Distribution')
df['area_sqft'].hist(ax=axes[0,1], bins=50)
axes[0,1].set_title('Area Distribution')
df['zone_type'].value_counts().plot(kind='bar', ax=axes[0,2])
axes[0,2].set_title('Zone Type Distribution')
df['dispute_risk'].value_counts().plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%')
axes[1,0].set_title('Dispute Risk Distribution')
df['infrastructure_score'].hist(ax=axes[1,1], bins=30)
axes[1,1].set_title('Infrastructure Score')
df['is_fraud'].value_counts().plot(kind='bar', ax=axes[1,2])
axes[1,2].set_title('Fraud Distribution')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['area_sqft', 'market_price', 'infrastructure_score', 'road_proximity_km', 'actual_value']
corr = df[numeric_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
