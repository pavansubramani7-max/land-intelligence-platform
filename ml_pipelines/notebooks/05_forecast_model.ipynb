# Price Forecast Model


```python
import sys
sys.path.insert(0, '../../backend')
from app.ml.train_forecast import train_forecast_models, generate_price_series, forecast_prices
import matplotlib.pyplot as plt

results = train_forecast_models('../data/sample_land_data.csv')
print(results)

series = generate_price_series(500000, n_periods=60)
forecast_df = forecast_prices(series, steps=36)

plt.figure(figsize=(12, 5))
series.plot(label='Historical', color='blue')
forecast_df['forecast'].plot(label='Forecast', color='orange')
plt.fill_between(forecast_df.index, forecast_df['lower'], forecast_df['upper'], alpha=0.2, color='orange')
plt.title('ARIMA Price Forecast')
plt.legend()
plt.show()
```
